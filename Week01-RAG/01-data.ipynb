```python
# import packages
import requests
```

```python
# common variables
URL_PREFIX='https://datatalks.club/faq'
```

```python
# fetch the courses
path = '/json/courses.json'
response = requests.get(url=f'{URL_PREFIX}{path}')
courses_raw = response.json()
```

```python
# for each course, extract the faq data and collate it into a single corpus of documents, with each document containing an id, course, section, question and answer
documents = []
for course in courses_raw:
    course_path = f'{URL_PREFIX}{course['path']}'
    course_response = requests.get(url=course_path)
    course_response.raise_for_status() # method to check whether HTTP request succeeded; if the server returns an error status code, it raises an exception; 
    # without this, the code execution might continue even though the request failed
    course_faqs = course_response.json()

    documents.extend(course_faqs)

print(f'Total number of faq documents: {len(documents)}')
```
