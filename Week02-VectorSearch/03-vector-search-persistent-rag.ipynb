```python
# import the required packages
from sentence_transformers import SentenceTransformer
from sqlitesearch import VectorSearchIndex
from openai import OpenAI
from rag_helper import RAGVector
import numpy as np
from dotenv import load_dotenv
import os
from pprint import pprint
```

```python
# load the environment variables
load_dotenv()

api_key = os.getenv('OPENAI_API_KEY')
hf_token = os.getenv('HF_TOKEN')

# specify the persistent storage paths
DB_DIR='.'
DB_NAME='faq.db'

# load the embedding model
model = SentenceTransformer(model_name_or_path='all-MiniLM-L6-v2', token=hf_token)
# initialize openai client
llm_client = OpenAI(api_key=api_key)
```

```python
# open the persistent index
vs_index = VectorSearchIndex(
    keyword_fields=['course'],
    mode='ivf',
    db_path=os.path.join(DB_DIR, DB_NAME)
)
```

```python
# test the search using persistent storage (without RAG) - sqlitesearch
# query = "I just discovered the course. Can I still join it?"
# query_vector = model.encode(query)

# results = vs_index.search(
#     query_vector,
#     filter_dict={'course':'llm-zoomcamp'},
#     num_results=5
# )

# pprint(results)
```

```python
# create a RAGVector object with the persistent vector index
vector_assistant = RAGVector(
    embedder=model,
    index=vs_index,
    llm_client=llm_client
)
```

```python
# test the rag assistant
query = 'the program has already begun, can I still sign up?'
answer = vector_assistant.rag(query=query)
print(answer)

# close the connection
vs_index.close()
```
